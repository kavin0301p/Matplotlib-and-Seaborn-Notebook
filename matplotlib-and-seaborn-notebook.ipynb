# **Plotting Multiple lines**

Here **range(lower_bound, upper_bound)** is used with respect to the number of entry in the lists watermelon and mango respectively 

```python
year = range(2014, 2024);
watermelon = [0.102, 0.983, 0.654, 0.873, 0.563, 0.985, 0.129, 0.522, 0.888, 0.549]
mango = [0.501, 0.859, 0.657, 0.678, 0.183, 0.152, 0.699, 0.999, 0.133, 0.876,]

plt.plot(year, watermelon);
plt.plot(year, mango);
plt.xlabel('Years');
plt.ylabel('Yield of various fruits(tons per hectares)');
```

# **Legend and Chart Title**

To differentiate between multiple plots, we use plt.legend([ ]) function. **Note that the multiple plots should be written as an element of a list**

To give title to the chart, we use plt.title() method.

```python
year_data = range(1990, 2000);
fig = [0.478, 0.983, 0.444, 0.234, 0.857, 0.578, 0.112, 0.385, 0.832, 0.771];
mosambi = [0.899, 0.123, 0.232, 0.456, 0.986, 0.999, 0.754, 0.328, 0.583, 0.228,];


plt.plot(year_data, fig);
plt.plot(year_data, mosambi);
plt.plot(year_data, watermelon);
plt.plot(year_data, mango);

plt.xlabel('Years');
plt.ylabel('Yield of various fruits');

plt.legend(['Fig', 'Mosambi', 'Watermelon', 'Mango']);
plt.title('Crop Yields of Chennai ');
```

# **Line Markers**

Line markers can be used to represent each data points by specific shapes during multiple plots. This can be done by including **marker** argument inside plt.plot() function. Matplotlib provides many marker options like diamond, circle, cross etc. It can be referred from -> https://matplotlib.org/3.1.1/api/markers_api.html

```python
year = range(2006, 2016);
plt.plot(year, watermelon, marker = 'o');
plt.plot(year, fig, marker = 'x');

plt.xlabel('Years');
plt.ylabel('Yield of various fruits');
plt.legend(['Watermelon', 'Fig']);
plt.title('Crop Yields in India');
```

# **Styling Lines and Markers**

The plt.plot() function supports many arguments for styling and markers ->
* **color or c** -> Selecting color for the graphline. Refer -> https://matplotlib.org/3.1.0/gallery/color/named_colors.html
* **linestyle or ls** -> Selecting solid or dashed line
* **linewidth or lw** -> Selecting the width of the line
* **markersize or ms** -> Selecting the size of the marker
* **markeredgecolor or mec** -> Selecting the edge color of the marker
* **markeredgewidth or mew** -> Selecting the edge width of the marker
* **markerfacecolor or mfc** -> Setting the fill colors of the marker
* **alpha** -> Used for opacity of the plot


```python
year = range(2006, 2016);

plt.plot(year, mango, marker = 's', c = 'g', ls = '--', lw = 3, ms = 3, mec = 'springgreen', mew = 2, mfc = 'blue');
plt.plot(year, watermelon, marker = 'D', c = 'r', ls = '-', lw = 2, ms = 2, mec = 'maroon', mew = 3, mfc = 'lightcoral', alpha = .2);

plt.xlabel('Years');
plt.ylabel('Yield of fruits');

plt.legend(['Mango', 'Watermelon']);
plt.title('Fruit crop yields in India');
```

# Note

The fmt argument is used as a shorthand for specifying **markershape, linestyle and linecolor**

**fmt = '[marker][line][color]'**

```python
years = range(2006, 2016);
plt.plot(years, fig, 'o-y');
plt.plot(years, watermelon, '*--b');

plt.xlabel('Years');
plt.ylabel('Yield of Various fruits');

plt.legend(['Fig', 'Watermelon']);
plt.title('Crop Yields in India');
```

# **SCATTER PLOT**

* In a Scatter plot, the values of two variables are plotted as points on a two-dimensional grid. Additionally, a third variable is used to determine the size or color of the points.
* Iris_dataset is a dataset present in seaborn library which contains info about petals and sepals of three varities of flowers. Since, this dataset is included in seaborn library, it can be loaded as a Pandas Dataframe

**Iris dataset -> https://en.wikipedia.org/wiki/Iris_flower_data_set**


```python
#Loading dataset to Pandas dataframe
import pandas as pd
flower_df = sns.load_dataset('iris');
flower_df
```

**As already mentioned this dataset contains info about three different varieties of flowers. But here only tow varities are visible -> Setosa and virginica. To know the different datas entered in the species column we use unique() method**

```python
flower_df.species.unique() 
```

**Let's plot a line chart with sepal_length and sepal_width to visualise the data**

```python
plt.plot(flower_df.sepal_length, flower_df.sepal_width);
```

**This line chart does not provide accurate info as there are too any combinations within the dataset.**

* So we use scatter plot from **Seaborn library** so that we get a scattered data points. Thus the visualisation of the graph will be easier than the line plot

```python
sns.scatterplot(x = flower_df.sepal_length, y = flower_df.sepal_width);
```

# **Adding Hues**

* In the above plot, it does not give any info about which flower has the more sepal_length and sepal_width. 
* So, we use **hues**, which is a marker that is present in line chart.
* Hues give different colours based on different categories. 
* **Also the name of the hue is the heading of the legend**
* To increase the size of the data points the variable **s** (size) can be increased or decreased.

```python
sns.scatterplot(x=flower_df.sepal_length, y=flower_df.sepal_width, hue=flower_df.species, s=100);
```

**Here we can easily identify that setosa has largest sepal_width while smallest sepal_length. Similarly, other species of flowers can be noticed**

# **Customising Seaborn Figures**

**Although, seaborn library uses matplotlib library internally, the title and the figure size of the plot can be adjusted using the matplotlib functions**

```python
sns.set_style('darkgrid')
plt.figure(figsize = (12, 6));
plt.title('Sepal dimensions of various varieties of flowers');

sns.scatterplot(x=flower_df.sepal_length, 
                y=flower_df.sepal_width , 
                hue=flower_df.species ,
                s=100);
```

# **Plotting using Pandas Dataframes**

* Seaborn has an inbuilt support for the pandas dataframe. 
* Instead of passing each series of a dataframe as variables for the plot, we can just include the column date and mention an argument **data** for specifing the name of the dataframe

```python
plt.figure(figsize=(12,6));
plt.title('Sepal dimensions');
sns.scatterplot(x='sepal_length', 
                y='sepal_width', 
                hue='species', 
                s=100, 
                data=flower_df);
```

# **Histogram**

A **Histogram** represents a bar graph consisting of various bins(cylinders) for visualising data. This can be done by using matplotlib library's function plt.hist()

Let's visualise **sepal_width** using histogram

```python
flowers_df = sns.load_dataset('iris');
flowers_df.sepal_width
```

```python
sns.set_style('darkgrid');

plt.title('Sepal width distribution');
plt.hist(flowers_df.sepal_width);
```

# **Note**

* Here we can note that the sepal width lies in the range 2.0-4.5.
* Y-Axis represents the number of sepals with a particular sepal_width -> **It's the count of sepals_width**
* We can observe that the maximum number of flowers (35 flowers) have sepal_width of the range 3.0-3.25

# **Controlling the size and the number of bins(cylinders)**

```python
#Mentioning the number of bins

plt.hist(flowers_df.sepal_width, bins=5);
```

```python
#Specifying the boundaries between two consecutive bins
import numpy as np

plt.hist(flowers_df.sepal_width, bins = np.arange(2,5,0.5)); #This will divide in equal intervals
```

```python
##Bins of unequal sizes
plt.hist(flowers_df.sepal_width, bins = [1,3,4,4.5]);
```

# **Multiple Histogram**

Similiar to line charts and scatter plots, we can create multiple histograms. During this, we will adjust the opacity so that all the plots are visible.

We will create separate dataframes for different varities of flower from flowers_df using **QUERING**

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
#Quering
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']

sns.set_style('darkgrid');
plt.title('Various distribution of flowers based on sepal_width');

plt.hist(setosa_df.sepal_width, alpha = 0.4, bins = np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width, alpha = 0.4, bins = np.arange(2,5,0.25));
plt.hist(virginica_df.sepal_width, alpha = 0.4, bins = np.arange(2,5,0.25));

plt.legend(['setosa', 'versicolor', 'virginica']);
```

**We can also stack multiple histogram layers on top of another**

```python
plt.title('Distribution of sepals');
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins = np.arange(2,5,0.25),
         stacked = True);
plt.legend(['setosa', 'versicolor', 'virginica']);
```

# **BAR CHART**

Bar charts are similiar to line charts. The only difference between two is the data points are connected through lines while in bar chart bar is used to display the data.
**We can use plt.bar() function to plot a bar chart**

```python
year_set = range(2010, 2016);
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]

plt.bar(year_set, apples);
```

**Like histograms, we can stack multiple bar**

So, we use bottom method inside plt.bar() to acheive this.

```python
plt.bar(year_set, apples);
plt.bar(year_set, oranges, bottom = apples);

plt.legend(['apples', 'oranges']);
```

# **Heatmap**
A heatmap is used to visualise data in two-dimensional matrix or table in form of colors. Let's use sample dataset from Seaborn, called **flights**, to visualize monthly passenger footfall at an airport over 12 years.

```python
flights_df = sns.load_dataset('flights').pivot(index='month', columns='year', values='passengers')
```

```python
flights_df #Now this is a matrix
```

**flights_df is a matrix with one row for each month and one column for each year. The values show the number of passengers (in thousands) that visited the airport in a specific month of a year. We can use the sns.heatmap function to visualize the footfall at the airport.**

```python
plt.title('No. Of Passengers');
sns.heatmap(flights_df);
```

* The brighter side indicates more number of values
* The darker side indicates less number of values.

**We can show the actual values in the block by annot=True argument and we can even change the color of the heatmap by cmap**

```python
sns.heatmap(flights_df,fmt='d', annot=True, cmap='Blues');
```

# **Plotting Multiple graphs in a grid**

* Both Matplotlib and seaborn supports multiple charts in a grid
* In Matplotlib,**plt.subplots() argument** is used which returns a set of axes for plotting
* Whenever we use seaborn library in multiple graphs **make sure to mention the axes -> ax=axes[0,1] etc**

```python
flowers_df
```

```python
fig, axes = plt.subplots(2,3,figsize=(20,8))  # 2 represnts the number of rows while 3 represents the number of columns

#GRAPH1
#PUsing axes for plotting
year = range(2000,2010);
mango = [0.234, 0.738, 0.984, 0.643, 0.298, 0.589, 0.398, 0.188, 0.998, 0.284]
fig = [0.739, 0.278, 0.489, 0.938, 0.989, 0.456, 0.344, 0.456, 0.789, 0.382]
axes[0,0].plot(year, mango, '*--b');
axes[0,0].plot(year, fig, '1-r');
axes[0,0].set_xlabel('Years');                      ##FOR LABEL SYNTAX CHANGES
axes[0,0].set_ylabel('Fruit Crop Yields(tons per hectares)');
axes[0,0].legend(['Mango', 'Fig']);
axes[0,0].set_title('Crop Yield in India');         ##FOR TITLE SYNTAX CHANGES
 
#GRAPH2
#Pass the axes into seaborn
axes[0,1].set_title('Petal Length vs Petal Width');
sns.scatterplot(x='petal_length', y='petal_width', ax=axes[0,1], data=flowers_df, s=100, hue='species');

##GRAPH3
axes[0,2].set_title('Distribution of Sepal Width');
axes[0,2].hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width], bins=np.arange(2,5,0.25), stacked=True);
axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);


##GRAPH4
axes[1,0].set_title('Restaurant bill');
sns.barplot(x='day', y='total_bill', data=tips_df, ax=axes[1,0]);

##GRAPH5
axes[1,1].set_title('FLIGHT TRAFFIC');
sns.heatmap(flights_df, cmap='Blues', ax=axes[1,1]);

##GRAPH6
```
